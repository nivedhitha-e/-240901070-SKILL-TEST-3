```python
# House Price Prediction (Linear Regression)

import pandas as pd
from sklearn.linear_model import LinearRegression

data2 = {
    'Year': [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019],
    'Price (₹ Lakhs)': [25,28,30,34,37,40,43,45,47,50]
}

df2 = pd.DataFrame(data2)

X2 = df2[['Year']]
y2 = df2['Price (₹ Lakhs)']

lr2 = LinearRegression()
lr2.fit(X2, y2)

test_year = pd.DataFrame([[2020]], columns=X2.columns)
pred2 = lr2.predict(test_year)

print(f"Predicted House Price for 2020: {pred2[0]:.2f} Lakhs")
```
