```python
4.Write a program to build a machine learning linear regression model to predict
the price of a house based on its features. You are given a dataset containing
information of Year and price.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
df = pd.DataFrame({
    'Year': [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019],
    'Price': [25,28,30,34,37,40,43,45,47,50]
})
X = df[['Year']]
y = df['Price']
model = LinearRegression().fit(X, y)
future = pd.DataFrame([[2025]], columns=['Year'])
predicted_price = model.predict(future)
print("Predicted House Price in 2025:", predicted_price[0], "Lakhs")
plt.scatter(df['Year'], df['Price'])
plt.plot(df['Year'], model.predict(X))
plt.xlabel('Year')
plt.ylabel('Price (Lakhs)')
plt.title('House Price Prediction')
plt.show()
```
