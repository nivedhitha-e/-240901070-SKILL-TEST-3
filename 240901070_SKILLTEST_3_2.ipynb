```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder

df = pd.DataFrame({
    "Voltage":[5,5.5,4.8,6,6.5,6.2,4.9,5.8,5.2,6.3],
    "Current":[0.5,0.4,0.6,0.3,0.2,0.25,0.55,0.35,0.5,0.22],
    "Status":["Working","Working","Working","Defective","Defective",
              "Defective","Working","Defective","Working","Defective"]
})

df["Status"] = LabelEncoder().fit_transform(df["Status"])

model = KNeighborsClassifier(n_neighbors=3).fit(
    df[["Voltage","Current"]], df["Status"]
)

sample = pd.DataFrame([[5.7,0.3]], columns=["Voltage","Current"])

print("Prediction:", "Working" if model.predict(sample)[0]==0 else "Defective")
```
