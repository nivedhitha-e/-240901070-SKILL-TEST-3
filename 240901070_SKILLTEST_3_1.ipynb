```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder

df = pd.DataFrame({
    "Amplitude":[1,1.5,0.8,2,1.2,1.7,0.9,2.1,1.1,1.8,0.95,2.2,1,1.9,1.3],
    "Frequency":[50,60,55,40,65,45,70,35,60,48,62,42,58,46,66],
    "Duration":[1,0.5,1.2,1,0.8,1.5,1,1.3,0.9,1,1.1,0.7,1,1.2,0.6],
    "Peak_to_Peak":[2,3,1.6,4,2.4,3.4,1.8,4.2,2.2,3.6,1.9,4.4,2,3.8,2.6],
    "Label":["Sine","Square"]*7 + ["Sine"]
})

df["Label"] = LabelEncoder().fit_transform(df["Label"])

model = KNeighborsClassifier(n_neighbors=3).fit(
    df[["Amplitude","Frequency","Duration","Peak_to_Peak"]],
    df["Label"]
)

# FIX: use DataFrame with same column names
sample = pd.DataFrame([[1.4,55,1,2.8]],
    columns=["Amplitude","Frequency","Duration","Peak_to_Peak"]
)

print("Prediction:", "Sine" if model.predict(sample)[0]==0 else "Square")
```
