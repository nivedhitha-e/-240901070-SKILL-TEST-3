```python
2. Write a program to classify a component is defective or not based on Voltage and
Current using KNN. Encode labels (Working → 1, Defective → 1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
df = pd.DataFrame({
    'Voltage': [5,5.5,4.8,6,6.5,6.2,4.9,5.8,5.2,6.3],
    'Current': [0.5,0.4,0.6,0.3,0.2,0.25,0.55,0.35,0.5,0.22],
    'Status': ['Working','Working','Working','Defective','Defective',
               'Defective','Working','Defective','Working','Defective']
})
df['Status'] = df['Status'].map({'Working': 0, 'Defective': 1})
X = df[['Voltage','Current']]
y = df['Status']
knn = KNeighborsClassifier(n_neighbors=3).fit(X, y)
new = pd.DataFrame([[5.7, 0.3]], columns=['Voltage','Current'])
print("Prediction:", "Defective" if knn.predict(new)[0] else "Working")
plt.scatter(df[df['Status']==0]['Voltage'], df[df['Status']==0]['Current'], label='Working')
plt.scatter(df[df['Status']==1]['Voltage'], df[df['Status']==1]['Current'], label='Defective')
plt.xlabel('Voltage')
plt.ylabel('Current')
plt.title('KNN Classification')
plt.legend()
plt.show()
```
