```python
# Electronic Components Clustering (K-Means)

import pandas as pd
from sklearn.cluster import KMeans

data6 = {
    'Voltage': [4.819941,5.249786,4.771151,4.602712,4.814919,
                12.38281,12.36406,11.22958,12.17975,12.0638],
    'Current': [0.87293,0.918358,0.958775,1.087788,0.877739,
                2.257141,2.383657,1.89107,1.82507,2.29575]
}

df6 = pd.DataFrame(data6)

X6 = df6[['Voltage','Current']]

kmeans6 = KMeans(n_clusters=2, random_state=0)
df6['Cluster'] = kmeans6.fit_predict(X6)

print(df6)
```
