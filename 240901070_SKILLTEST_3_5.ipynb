```python
# Signal Quality Prediction (Decision Tree)

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

data5 = {
    'Transmitter': ['T1','T2','T1','T3','T2','T1','T3','T2','T1','T3'],
    'Signal_Strength': [75,50,90,40,85,70,30,60,80,20],
    'Frequency': [2000,1500,2200,1300,2100,1800,1200,1600,1900,1100],
    'Signal_Quality': ['Good','Bad','Good','Bad','Good','Good','Bad','Bad','Good','Bad']
}

df5 = pd.DataFrame(data5)

# Encoding
le1 = LabelEncoder()
df5['Transmitter'] = le1.fit_transform(df5['Transmitter'])

le2 = LabelEncoder()
df5['Signal_Quality'] = le2.fit_transform(df5['Signal_Quality'])

X5 = df5[['Transmitter','Signal_Strength','Frequency']]
y5 = df5['Signal_Quality']

dt5 = DecisionTreeClassifier()
dt5.fit(X5, y5)

test_data = pd.DataFrame([[0, 78, 2000]], columns=X5.columns)  # T1 = 0
pred5 = dt5.predict(test_data)

print(f"Predicted Signal Quality: {'Good' if pred5[0]==1 else 'Bad'}")
```
